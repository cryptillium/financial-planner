## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs (endpoint)
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# Execute GET request for BTC and store response
btc_response_data = requests.get(btc_url)

# Formatting as json
btc_json_data = btc_response_data.json()

# accessing the main branch relevant to the properties for BTC
btc_data = btc_json_data["data"]["1"]

# extracting the price of BTC in USD
btc_price = btc_data["quotes"]["USD"]["price"]

# Fetch current ETH price
# Execute GET request for ETH and store response
eth_response_data = requests.get(eth_url)

# Formatting as json
eth_json_data = eth_response_data.json()

# accessing the main branch relevant to the properties for ETH
eth_data = eth_json_data["data"]["1027"]

# extracting the price of BTC in USD
eth_price = eth_data["quotes"]["USD"]["price"]

# Compute current value of my crpto
my_btc_value = btc_price * my_btc
my_eth_value = eth_price * my_eth

# Print current crypto wallet balance
print(f"The current value of my {my_btc} BTC is ${my_btc_value:0.2f} USD")
print(f"The current value of my {my_eth} ETH is ${my_eth_value:0.2f} USD")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
now = pd.Timestamp("2020-09-15", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
df_portfolio = alpaca.get_barset(
    tickers,
    timeframe,
    start = now,
    end = 1000,
    limit=1000
).df

# Preview DataFrame
df_portfolio
```

```python
# Pick AGG and SPY close prices

# Create an empty DataFrame for closing prices
df_closing_prices = pd.DataFrame()

# Fetch the closing prices of AGG and SPY
df_closing_prices["AGG"] = df_portfolio["AGG"]["close"]
df_closing_prices["SPY"] = df_portfolio["SPY"]["close"]

# Drop the time component of the date
df_closing_prices.index = df_closing_prices.index.date

# sort the dataframe so that the most recent price is first
df_closing_prices.sort_index(inplace=True, ascending=False)

# Display sample data
df_closing_prices.head(10)
```

```python
# retrieve current price for AGG
agg_close_price = df_closing_prices.iloc[0]["AGG"]

# retrieve current price for SPY
spy_close_price = df_closing_prices.iloc[0]["SPY"]

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# calculate AGG value
my_agg_value = agg_close_price * my_agg

# calculate SPY value
my_spy_value = spy_close_price * my_spy

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Consolidate financial assets data
# consolidating the value of assets together..
assets_data = {
    "amount": [my_btc_value + my_eth_value, my_agg_value + my_spy_value]
}

# Set the assets
assets = ["crypto", "shares"]

# Create savings DataFrame
df_savings = pd.DataFrame(assets_data, index=assets)

# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# Creating a pie chart to show the proportion of savings between each financial asset

df_savings.plot.pie(y="amount", title="Composition of Personal Savings")
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# creating a function to calculate total savings...

def get_savings():
    """
    The purpose of this function is to calculate the total savings
    get_savings() method loops through the amount column of df_savings dataframe
    and performs cumulative addition for each row to arrive at the total savings amount
  
    """
    # reset savings before beginning calculation..
    savings = 0
    
    # iterate through the amount column
    for column_name, item in df_savings["amount"].iteritems():
        # incrementally add each item
        savings+=item
        
    # when done, return the total    
    return savings

# get total savings..
total_savings = get_savings()
print(f"Total Savings = ${total_savings}")
```

```python
# Validate saving health
# creating a function to validate health savings ...

def validate_health_savings(savings, emergency):
    """
    This function returns the health savings status.
    
    input parameters:
        savings:     This is the total amount of savings        - mandatory
        emergency:   This is the total emergency fund available - mandatory
        
    function design logic:
    
        (1) If total savings are greater than the emergency fund, display a message congratulating the person for having enough money in this fund.

        (2) If total savings are equal to the emergency fund, display a message congratulating the person on reaching this financial goal.

        (3) If total savings are less than the emergency fund, display a message showing how many dollars away the person is from reaching the goal.    
    """
    
    if savings > emergency:
        return "congratulations, you have enough money in this fund"
    elif savings == emergency:
        return "congratulations, you have reached your financial goal for this fund"
    else:
        return f"you are ${emergency - savings} away from reaching your goal for this fund"

#validate health savings fund - determine if we have enough
print(validate_health_savings(total_savings, emergency_fund))    
    
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-09-15', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-09-15', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)

# retrieving historical data for AGG and SPY for the last years from today
df_stock_data = alpaca.get_barset(
    tickers,
    timeframe,
    start = start_date,
    end = end_date,
    limit=1000
).df




# Display sample data
df_stock_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30year = MCSimulation(
    portfolio_data = df_stock_data,
    weights = [.40,.60],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
MC_30year.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30year.calc_cumulative_return()
```

```python
# Plot simulation outcomes for 30 year simulation
line_plot_30year = MC_30year.plot_simulation()

# Save the plot for future usage
line_plot_30year.get_figure().savefig("MC_30year_sim_plot.png", bbox_inches="tight")
```

```python
# Plot probability distribution and confidence intervals for 30 year simulation
dist_plot_30year = MC_30year.plot_distribution()

# Save the plot for future usage
dist_plot_30year.get_figure().savefig('MC_30year_dist_plot.png',bbox_inches='tight')
```

### Retirement Analysis

```python
# Fetch summary statistics from the 30 year Monte Carlo simulation results
summary_stats_30year = MC_30year.summarize_cumulative_return()

# Print summary statistics
print(summary_stats_30year)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(summary_stats_30year[8]*initial_investment,2)
ci_upper = round(summary_stats_30year[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(summary_stats_30year[8]*initial_investment,2)
ci_upper = round(summary_stats_30year[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
